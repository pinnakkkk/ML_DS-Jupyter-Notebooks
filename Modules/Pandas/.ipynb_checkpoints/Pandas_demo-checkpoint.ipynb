```python
colors_duration_df = df[['color','duration']]
colors_duration_df.head()
```

```python
colors_duration_df.loc[0:,'duration'] = colors_duration_df['duration'].str.replace(" ",":")
colors_duration_df['duration'] = colors_duration_df['duration']+':00'
colors_duration_df
```

```python
colors_duration_df['duration'] = pd.to_timedelta(colors_duration_df['duration'])
```

```python
colors_duration_df.info()
colors_duration_df
```

```python
colors_duration_df.loc[colors_duration_df['color'] == 'Amber', 'color'] = 1
colors_duration_df.loc[colors_duration_df['color'] == 'Red', 'color'] = 2
colors_duration_df.loc[colors_duration_df['color'] == 'Black','color'] = 3
colors_duration_df
```

```python
plt.scatter(colors_duration_df['color'], colors_duration_df['duration'])
plt.show()
```
